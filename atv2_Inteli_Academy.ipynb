**Resposta =** Estudantes que dormem menos de 6 horas por noite tendem a ter notas finais menores que estudantes que dormem mais de 6 horas..



```python
# Criar grupos
grupo_menos_6h = df[df['sleep_hours'] < 6]['exam_score']
grupo_mais_6h = df[df['sleep_hours'] >= 6]['exam_score']

# Calcular médias
media_menos_6h = grupo_menos_6h.mean()
media_mais_6h = grupo_mais_6h.mean()

# Plotar gráfico
plt.bar(['menos que 6 horas de sono', 'maior ou igual a 6 horas de sono'], [media_menos_6h, media_mais_6h], color=['red', 'green'])
plt.ylabel('Média da nota final')
plt.title('Média da nota final por grupo de horas de sono')
plt.show()
```
